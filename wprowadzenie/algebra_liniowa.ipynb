# Algebra liniowa

```python
import numpy as np
```

Norma wektora, długośc wektora w R^2image.png

```python
v1= np.array([-1,4])
v1
```

Obliczanie normy wektora

```python
np.linalg.norm(v1)
```

Norma euklidesowa image.png

```python
v2 = np.array([-10,10,13])
np.linalg.norm(v2)
```

Odległość dwóch punktó na płaszczyźnie

```python
p = np.array([3,0])
q = np.array([0,4])
np.linalg.norm(p-q)
```

Odległośc dwóch punktów w przestrzenui R^n image.png

```python
p = np.array([3,0,1])
q = np.array([0,4,2])
np.linalg.norm(p-q)
```
