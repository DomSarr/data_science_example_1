# Rozwiązywanie ukłądów równań

```python
A = np.array([[2,4],[1,-1]])
A
```

```python
B = np.array([[10],[-1]])
B
```

```python
A_inv = np.linalg.inv(A)
X = np.dot(A_inv,B)
X
```
